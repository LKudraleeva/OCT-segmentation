```python
paths = ['2015_BOE_Chiu/Subject_01.mat', '2015_BOE_Chiu/Subject_02.mat', '2015_BOE_Chiu/Subject_03.mat',
         '2015_BOE_Chiu/Subject_04.mat', '2015_BOE_Chiu/Subject_05.mat', '2015_BOE_Chiu/Subject_06.mat',
         '2015_BOE_Chiu/Subject_07.mat', '2015_BOE_Chiu/Subject_08.mat', '2015_BOE_Chiu/Subject_09.mat',
         '2015_BOE_Chiu/Subject_10.mat']
keys = ['automaticLayersDME','automaticLayers2','manualLayers1', 'manualLayers2']
```

```python
import imageio
import scipy.io
import numpy as np
```

```python
for num, p in enumerate(paths):
    data = scipy.io.loadmat(p) # загружаем данные

    # определяем индексы размеченных изображений
    ims = data['manualLayers1']
    indexes = []
    for i in range(61):
        if np.isnan(ims[:, :, i]).all():
            continue
        else:
            indexes.append(i)

    # исхдные изображения
    images = data['images']

    # для каждого размеченного изображения определяем маски слоев
    for index in indexes:

        # сохраняем исходное изображение
        imageio.imsave('Images/' +str(num)+ '_'+str(index)+ '.png', images[:,:,index])

        for k in keys:
            df = data[k]
            im = df[:, :, index] # разметка
            first = im[:, 118:654]
            first=first.astype(int)
            layers = []
            for i in range(0,7):
                l = np.zeros((496, 768), dtype=int)
                for j in range(0, 536):
                    if i + 2 < 7:
                        if first[i+1][j] < 0:
                            if first[i+2][j] > 0:
                                first[i+1][j] = (first[i+2][j] + first[i][j]) // 2
                    if first[i][j] > 0 and first[i+1][j] > 0:
                        for m in range(first[i][j], first[i+1][j]):
                                l[m][j+118]=1
                layers.append(l)
                imageio.imsave('Labels/' + k + '/'+str(num)+ '_' + str(index) + '_'+str(i)+'.png', layers[i])
```

```python
paths = ['2015_BOE_Chiu/Subject_01.mat', '2015_BOE_Chiu/Subject_02.mat', '2015_BOE_Chiu/Subject_03.mat',
         '2015_BOE_Chiu/Subject_04.mat', '2015_BOE_Chiu/Subject_05.mat', '2015_BOE_Chiu/Subject_06.mat',
         '2015_BOE_Chiu/Subject_07.mat', '2015_BOE_Chiu/Subject_08.mat', '2015_BOE_Chiu/Subject_09.mat',
         '2015_BOE_Chiu/Subject_10.mat']
keys = ['manualLayers1', 'manualLayers2']
# keys = ['automaticLayersDME','automaticLayers2','manualLayers1', 'manualLayers2']
```

```python
for num, p in enumerate(paths):
    data = scipy.io.loadmat(p) # загружаем данные

    # определяем индексы размеченных изображений
    ims = data['manualLayers1']
    indexes = []
    for i in range(61):
        if np.isnan(ims[:, :, i]).all():
            continue
        else:
            indexes.append(i)

    # исхдные изображения
    images = data['images']
    labels = []
    # для каждого размеченного изображения определяем маски слоев
    for index in indexes:

        for k in keys:
            df = data[k]
            im = df[:, :, index] # разметка
            first = im[:, 118:654]
            first=first.astype(int)

            l = np.zeros((496,768), dtype=int)
            for j in range(0,536):
                for i in range(0,7):
                    if first[i][j]<0:
                        first[i][j] = first[i][j-1]
                    else:
                        if first[i+1][j]>0:
                            for m in range(first[i][j], first[i+1][j]):
                                l[m][j+118]=i+1
                        elif first[i+2][j]>0:
                            for m in range(first[i][j], first[i+2][j]):
                                l[m][j+118]=i+1
                        elif first[i+3][j]>0:
                            for m in range(first[i][j], first[i+3][j]):
                                l[m][j+118]=i+1
                        elif first[i+4][j]>0:
                            for m in range(first[i][j], first[i+4][j]):
                                l[m][j+118]=i+1
                        elif first[i+5][j]>0:
                            for m in range(first[i][j], first[i+5][j]):
                                l[m][j+118]=i+1
                        elif first[i+6][j]>0:
                            for m in range(first[i][j], first[i+6][j]):
                                l[m][j+118]=i+1

            imageio.imsave('y/' + k + '/'+str(num)+ '_' + str(index) +'.png', l)
```

```python
from keras import backend as K

# Функция, которая смотрит на пересечение областей. Нужна для accuracy
def dice_coef(y_true, y_pred):
    return (2. * K.sum(y_true * y_pred) + 1.) / (K.sum(y_true) + K.sum(y_pred) + 1.)
```
